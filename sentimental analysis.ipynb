```python
import pandas as pd
```

```python
data = pd.read_csv('C:\\Users\\Sai\\Desktop\\ML\\Training Data (2).csv')
```

```python
data1 = data.iloc[:,3:5]
```

```python
import re
#Empty list
text1 = []
#Here we are removing the domain address
url_reg  = r'[a-z]*[:.]+\S+'
for i in data['text']:
     text1.append(re.sub(url_reg, '', i))
```

```python
#Converting into Data Frame
text1 = pd.DataFrame(text1,columns = ['text1'])
```

```python
#Concatinatin the two data frames
dataset = pd.concat([data1,text1], axis = 1)
dataset.head()
```

```python
#Here removing the uncleaned Text column
dataset.drop(['text'],inplace = True,axis =1)
```

```python
dataset = dataset[['text1','tone']]
```

```python
lower = []
for i in dataset['text1']:
    z = i.lower()
    lower.append(z)
```

```python
dataset1 = dataset1.values
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
clf = LabelEncoder()
dataset1[:,1] = clf.fit_transform(dataset1[:,1])
```

```python
data = pd.DataFrame(dataset1, columns = ['text','tone'])
```

```python
from textblob import TextBlob
```

```python
list = []
for i in data.text:
    obj = TextBlob(i)
    list.append(obj)
```

```python
list[0:5]
```

```python
senti = []
for i in list:
    sentiment = i.sentiment.polarity
    print(sentiment)
    senti.append(sentiment)
```

```python
sentiment = pd.DataFrame(senti,columns = ['sentiment'])
```

```python
sentiment.head()
```

```python
df = pd.concat([data,sentiment],axis = 1)
```

```python
senti = []
for i in df.sentiment:
    if i > 0.4:
        senti.append('positive')
    else:
        senti.append('negative')
```

```python
sentim = pd.DataFrame(senti,columns = ['sentiment_text'])
```

```python
final_data = pd.concat([df,sentim],axis = 1)
```
